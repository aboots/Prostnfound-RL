```python
import pandas as pd 
import wandb
api = wandb.Api()

runs = api.runs("pfrwilson/miccai2024_comparison")
info = []

for run in runs: 
    info_i = {}
    info_i['name'] = run.name 
    info_i['group'] = run.group
    info_i.update(run.summary._json_dict)
    info.append(info_i)
    
runs_df = pd.DataFrame(info)
```

```python
group = 'sam_ft'
epoch = 13
runs = api.runs(
    "pfrwilson/miccai2024_comparison", 
    filters={
        "group": group
    }
)

relevant_metrics = [
    'val/core_auc',
    'val/sens_at_20_spe',
    'val/sens_at_40_spe',
    'val/sens_at_60_spe',
    'val/bal_prop_pred_err'
]

data = []
for run in runs: 
    print(run.name)
    hist = run.history(
        keys=['epoch'] + relevant_metrics, 
    )
    data.append(
        dict(**hist.loc[hist['epoch'] == epoch].iloc[0].to_dict(), 
        group=run.group)
    )
df = pd.DataFrame(data)

df.groupby('group')[relevant_metrics].aggregate(['mean', 'std'])
```
